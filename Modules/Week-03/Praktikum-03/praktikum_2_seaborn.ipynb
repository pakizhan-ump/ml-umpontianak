```python
import seaborn as sns
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

1. Basic Seaborn Setup dan Style

```python
print("=== SEABORN SETUP DAN STYLE ===")

# Set style seaborn
sns.set_style("whitegrid")
sns.set_palette("Spectral")
```

2. Distribution Plots

```python
print("=== DISTRIBUTION PLOTS ===")

# Create sample data
np.random.seed(42)
data = np.random.normal(0, 1, 1000)

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Histogram dengan KDE
sns.histplot(data=data, kde=True, ax=axes[0,0])
axes[0,0].set_title('Histogram dengan KDE')

# KDE Plot
sns.kdeplot(data=data, ax=axes[0,1], fill=True)
axes[0,1].set_title('KDE Plot')

# Box Plot
sns.boxplot(data=data, ax=axes[1,0])
axes[1,0].set_title('Box Plot')

# Violin Plot
sns.violinplot(data=data, ax=axes[1,1])
axes[1,1].set_title('Violin Plot')

plt.tight_layout()
plt.show()
```

3. Categorical Plots

```python
# Load sample dataset
tips = sns.load_dataset("tips")
print("Tips dataset shape:", tips.shape)
print("\nFirst 5 rows:")
print(tips.head())
```

```python
print("=== CATEGORICAL PLOTS ===")

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Bar plot
sns.barplot(data=tips, x='day', y='total_bill', ax=axes[0,0])
axes[0,0].set_title('Bar Plot - Total Bill by Day')

# Box plot
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', ax=axes[0,1])
axes[0,1].set_title('Box Plot - Total Bill by Day and Sex')

# Violin plot
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', 
               split=True, ax=axes[1,0])
axes[1,0].set_title('Violin Plot - Total Bill by Day and Sex')

# Swarm plot
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', 
              size=3, ax=axes[1,1])
axes[1,1].set_title('Swarm Plot - Total Bill by Day and Sex')

plt.tight_layout()
plt.show()
```

4. Relational Plots

```python
print("=== RELATIONAL PLOTS ===")

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Scatter plot
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', 
                size='size', sizes=(20, 200), ax=axes[0,0])
axes[0,0].set_title('Scatter Plot - Tip vs Total Bill')

# Line plot
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', 
             err_style='band', ax=axes[0,1])
axes[0,1].set_title('Line Plot - Tip vs Total Bill')

# Relational plot dengan regression
sns.regplot(data=tips, x='total_bill', y='tip', ax=axes[1,0])
axes[1,0].set_title('Reg Plot - Tip vs Total Bill')

# LM plot dengan hue
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker', 
           col='time', height=5, aspect=1.2)
plt.suptitle('LM Plot - Tip vs Total Bill by Smoker and Time', y=1.02)
plt.show()
```

5. Matrix Plots

```python
# Load sample dataset
tips = sns.load_dataset("tips")
print("Tips dataset shape:", tips.shape)
print("\nFirst 5 rows:")
print(tips.head())
```

```python
# In[10]: Matrix Plots
print("=== MATRIX PLOTS ===")

# Select only numerical columns for correlation
numerical_tips = tips.select_dtypes(include=[np.number])

plt.figure(figsize=(6, 5))

# Correlation heatmap
correlation_matrix = numerical_tips.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)

plt.suptitle('Correlation Heatmap')
plt.show()

# Pair plot (akan dibuat di figure terpisah)
plt.figure(figsize=(10, 8))
sns.pairplot(tips, hue='time', diag_kind='hist', palette='viridis')
plt.suptitle('Pair Plot of Tips Dataset')
plt.show()
```
